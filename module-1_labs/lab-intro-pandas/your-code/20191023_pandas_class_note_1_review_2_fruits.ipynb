```python
import pandas as pd # import python library pandas as "pd"
import numpy as np # import python library numpy as "np"
```

```python
col_names = ["banana_density", "banana_mass", "apple_density", "apple_mass"]
# define four column names as "col_names" for float data sampled by numpy
df = pd.DataFrame(np.random.random_sample((10, 4)), columns = col_names)
# sampling 10 rows and 4 columns float data by numpy using column defined as col_names & define "df" 
```

```python
df # check a 10 by 4 dataframe sampled in above command line
```

```python
apples = ["apple_density", "apple_mass", "apple_mass"]
# rearrange two columns and add(copy) a column except two columns & define as "apples"
df[apples]
# confirm dataframe defined as "apples"
```

```python
df[["apple_density", "apple_mass", "apple_mass"]]
# confirm dataframe defined as "apples" using different method
```

```python
df.sort_values("banana_density", ascending = False)
# sort dataframe based on "banana_density" in descending order
# "ascending = False" means desceding
```
